**Analysis-*

*Central Tendency*

1. Numerical features like Length_url, page_rank, web_traffic etc. were analyzed for mean, median, standard deviation
2. Numerical features like length_url, length_hostname, ip showed nearly symmetric distribution meaning that the mean is approimately equal to the median

*Dispersion*

1. Binary features in the dataset like dns_record etc. have median between -1 and 1
2. Binary features in the dataset have lower variance(Lower Dispersion) as compared to the numerical features which have been found to have higher variance(Higher Dispersion)
3. Binary features are more stable as compared to the Numerical features

*Distribution*

1. Binary features aren discrete having values of -1,0 and 1. One class might be more frequent than the others leading to imbalance.
2. Some features have values of -1 which may indicate negative or missing values and needs to be handled
